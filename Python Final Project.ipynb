```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
import seaborn as sns 
import csv
import scipy.optimize as fit 
```

```python
df = pd.read_csv("PS_2023.11.27_15.28.29.csv",low_memory = False)  
dfadone = df.dropna(axis = 0, how = 'any') 

lowerparameter = dfadone[dfadone.pl_eqt >= 255]
finalwparam = lowerparameter[lowerparameter.pl_eqt <=300]
finalwparam 

```

```python
xdata = finalwparam.st_lum 
ydata = finalwparam.pl_eqt 

plt.figure(figsize = (10,7))
plt.scatter(xdata,ydata,color = "blue", marker = "D" )
plt.title("Golidilocks Planets(With One Sun) With Respect to Equillibrium Temperature and their Sun's Luminosity")
plt.xlabel("Solar Luminosity")
plt.ylabel("Planetary Equilibrium Temperature (Kelvins)")
plt.show() 
```

```python
def f(x, m, b): 
    return m * x + b

p0 = [0.5, 2]
parameters, covariance_matrix = fit.curve_fit(f, xdata, ydata, p0)

m = parameters[0] 
b = parameters[1] 
m_error = (covariance_matrix[0,0]) **.5 
b_error = (covariance_matrix[1,1]) **.5 

x = np.linspace(x.min(), x.max(), 100)
y = f(x, m, b)
plt.figure(figsize = (10,7)) 
plt.plot(x, y, color = "magenta", linewidth = "5")
plt.scatter(xdata, ydata, color = "blue", marker = "D") 

plt.title("Relationship between Solar Luminosity and Goldilocks Planet's Equilibrium Temperature")
plt.xlabel("Solar Luminosity")
plt.ylabel("Planetary Equilibrium Temperature (Kelvins)")
plt.show() 
  

```
